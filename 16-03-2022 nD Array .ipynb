## .reshape() : 
#### Reshaping means changing the shape of an array.
#### The shape of an array is the number of elements in each dimension.
#### By reshaping we can add or remove dimensions or change number of elements in each dimension

```python
x= np.array([[1,2,3,4],[5,6,7,8]])
print(x)
```

```python
x.shape,x.ndim,x.dtype
```

### Converting x from 2 D Array to 1 D Array

```python
x.reshape(8)
```

```python
x.reshape(8,1)
```

```python
x.reshape(1,8)
```

```python
x=x.reshape(2,4)
print(x)
```

```python
x.shape,x.ndim,x.dtype
```

###  numpy.transpose(a, axes=None)
### Reverse or permute the axes of an array; returns the modified array.
### For an array a with two axes, transpose(a) gives the matrix transpose.

```python
x_t = np.transpose(x)
x_t
```

```python
x.dot(x_t)
```

```python
I = np.identity(3)
I
```

### linspace() function returns number spaces evenly w.r.t interval. Similar to numpy.arange() function but instead of step it uses sample number. 

```python
np.linspace(0,1,24)
```

```python
#It is 3D Array
c = np.linspace(0,1,24).reshape(3,2,4)
c
```

```python
c.ndim,c.shape
```

```python
 #4D array
d = np.linspace(0,1,72).reshape(3,2,4,3)
d
```

```python
d.ndim,d.shape
```

```python
#5d Array
e = np.linspace(0,1,144).reshape(2,3,2,4,3)
e
```

```python
e.ndim,e.shape
```

### Slicing with 2D array
#### Slicing in python means taking elements from one given index to another given id
#### [start:end:step]

```python
x = np.arange(1,25).reshape(6,4)
x
```

```python
x[:2,-2:]
```

```python
x[2:-1,1:-1]
```

```python
#Print Last 2 columns and last 2 rows
x[-2:,-2:]
```
