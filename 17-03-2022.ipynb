### Seed: Generates the same random Numbers whenever it is reruned or 
repeated
Here there is no need to store the data in a variable
Whenever we repeat the same random number

```python
np.random.seed(1)
print(np.sum(np.random.random_sample(20)>0.5))
np.random.seed(1)
print(np.sum(np.random.random_sample(20)<=0.5))
```

```python
np.random.seed(2)
print(np.sum(np.random.random_sample(20)>0.5))
np.random.seed(2)
print(np.sum(np.random.random_sample(20)<=0.5))
```

```python
#Here same values are printed
np.random.seed(45)
np.random.random_sample(20)
```

```python
np.random.seed(45)
np.random.random_sample(20)>=0.5
```

```python
np.random.seed(45)
np.sum(np.random.random_sample(20)>=0.5)
```

### randint() method returns an integer number selected element from the specified range.
### To create a matrix of random integers in Python, randint() function of the numpy module is used. This function is used for random sampling i.e. all the numbers generated will be at random and cannot be predicted at hand.

```python
np.random.randint(1,100,10)
```

```python
np.random.randint(1,100,(10,4))
```

```python
#3D array
np.random.randint(1,100,(10,4,3))
```

#### We can have N dimensions of Array using randint

```python
np.random.randint(1,50,(2,5,6,7,4,6,1)).ndim
```

```python
np.random.randint(1,50,(2,5,6,7,4,6,1)).shape
```
